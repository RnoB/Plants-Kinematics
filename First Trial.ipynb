# Universal Plant Simulator

```python
import plantsKin as pk
from math import pi
import numpy as np
import matplotlib.pyplot as plt
```

## Generate a Single Plant

```python
plant = []
plant = pk.Plant(x0 = [0,0,0],theta0 = pi/2,N = 1000,dt=.01,growth = 'Apical',growthRate=1)
```

### Add tropism

```python
plant.addInteractions(name = 'ApicalTropism' ,intensity=-10,direction = 0)
plant.addInteractions(name = 'Tropism' ,intensity=-10,direction = pi/3)
plant.addInteractions(name = 'Proprioception' ,intensity=-.1)
```

```python
fig = plt.figure(figsize=(3, 3), dpi=100)
ax = fig.add_subplot(111)
for k in range(0,100):
    ax.plot(plant.x[:,0],plant.x[:,2])
    plant.update()
```

## Generate a Roots System

```python
roots = pk.Roots(4,.5,theta0 =pi,growth='Exponential')
```

### Add Tropism

```python
plant.addInteractions(name = 'ApicalTropism' ,intensity=-10,direction = 0)
plant.addInteractions(name = 'Tropism' ,intensity=-10,direction = pi/3)
plant.addInteractions(name = 'Proprioception' ,intensity=-.1)
```

### Add Collective Tropism

```python
roots.addCollectiveInteraction(name ='Apical')
```

```python
fig = plt.figure(figsize=(10, 10), dpi=100)
ax = fig.add_subplot(111)
for k in range(0,10):
    for root in roots.roots:
        ax.plot(root.x[:,0],root.x[:,2])
    roots.update()
```
